```python
# all imports:
import configparser,twitter
import pandas as pd
```

```python
# api keys are in config.ini to keep them outside of this public notebook

config = configparser.ConfigParser()
config.read('config.ini')
for key in config['DEFAULT']:  
    print(key)
twit = config['DEFAULT']
```

```python
#OAuth authentication 
#this is the new mode of authenticatino in tweet
#The consumer key 
consumer_key = twit["c_key"]
consumer_secret = twit["c_secret"]
#the access tokens can be found on application details
#page located at https://dev.twitter.com/apps (under "OAuth settings")
access_token = twit["a_token"] 
access_token_secret = twit["a_secret"]


## create the api object with the twitter-python library
api = twitter.Api(consumer_key=consumer_key, consumer_secret=consumer_secret, access_token_key=access_token, access_token_secret=access_token_secret)
#Test the api object method GetUSerTimeline with user screen_name='Seif_mej' which is my account to see all status
print(api.GetUserTimeline(screen_name='Seif_mej'))
```

```python
status = api.GetUserTimeline(screen_name='Seif_mej')
#Extract the text specific field from the status 
texts=[s.text for s in status]
#Get all following/followers in my account
users=api.GetFriends()
#Extract all the following friends names 
friends=[f.name for f in users]
#print 
print(users)

status=api.GetSearch(term ="Earthquake",count =2)
search_res=[s for s in status]
print(search_res)
```

```python
## get the user timeline with screen_name = 'stat110'
twitter_statuses = api.GetUserTimeline(screen_name = 'Seif_mej')

## create a data frame
## first get a list of panda Series or dict
pdSeriesList = [pd.Series(t.AsDict()) for t in twitter_statuses]

## then create the data frame
data = pd.DataFrame(pdSeriesList)

data.head(1)
# print(data.retweet_count)
```

```python
#Filter tweets with enough retweet count
maybe_interesting=data[data['retweet_count']>10]
maybe_interesting
```

```python
#get the text for each of the interesting tweets that counts more than 10 retweets
text_tweets=maybe_interesting.text
##print them out 

for t in text_tweets.values:
    print('---'*3)
    print(t)
```
